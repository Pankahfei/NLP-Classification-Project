## Part 3: Hyperparameter tuning 

####  This is a classification problem, there is 2 classes in this project which is `MarvelStudio` and `DC Cinematic`.

The first session will be preprocessing steps - lemmatisation.
Then, the feature column will be transformed into sparse matrix using Count Vectoriser or TfidfVectorizer
- CountVectorizer is used to transform a given text into a vector on the basis of the frequency (count) of each word that occurs in the entire text.
- TFIDF indicates how important a specific term in a document by reduce the weight of a term if the term's occurrences are scattered throughout all the documents

This two preprocessing method combine with each model will be compared using pipeline in gridsearching with respective accuracy score.
The most suited preprocessing method is used in building our final model.

Next, use randomized searchCV to has a simple hyperparameter tune on the models.
In this session, multiple models from base model to assemble model will be put into tuning.
1. Logistic Regression
2. K Neighbors Classifier
3. Decision Tree Classifier
4. Support Vector Classifier
5. Multinomial Naive Bayes Classifier
6. Random Forest Classifier
7. Stacking (Voting)
8. Bagging
9. Boosting (ADA Boost and Gradient Boost)
Then the model with good scores and generalisation will be choosen and further tune to get the best overall model

In the last session of this notebook, the best models are further tuned using Gridsearch CV. This searching method run larger iteration(longer runtime) but might able to get better score and close the generalistion gap. The trade off between the score and run time will be evaluate too.

### Proprocessing of data

```python
# the default NLTK stopword list
stop_words = set(stopwords.words('english'))  

# add additional stopwords
additional_stopwords = {'like','think','just','new'}
stop_words = stop_words.union(additional_stopwords)
```

```python
# lemmatise and remove stopwords from feature
def lemmatise(text):
    
   # Split and lemmatize words
    words = text.split(" ")
    lemmatizer = WordNetLemmatizer()
    words_lem = [lemmatizer.lemmatize(i) for i in words]

    # Join the words back into one string separated by space, 
    # and return the result.
    words=[]
    for i in (words_lem):   
        if i != "":
            words.append(i)
    
    no_stop_words = [token for token in words if token not in stop_words]

    return (' '.join(no_stop_words))
 
```

```python
X['title_combined']= X['title_combined'].apply(lambda x: lemmatise(x))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    random_state=42,
                                                    stratify=y)
```

### Hyperparameter Tuning - RandomSearchCV

Select the few model with better performance and generalised well from this session
Usinf 1000 iteration and 3 cross validation, sort out the optimized hyperparamter range for each model to be further tuned in next section.

Count vectorizer or Tfidf vectorizer pair with each model are hyperparameter tuned too in the function created below. At the end of this section, either Count Vectorizer or Tfidf will be used in our final model base on which pair having better performance. 


```python
# Parameter Grid for base model

#Multinomial NB
params_ran_nb = {
    'alpha':[1,0.9,0.8,0.7,0.6]
    
}

#Decision Tree
params_ran_dt = {
    'max_depth' : np.arange(1,6,1),
    'min_samples_split': np.arange(2,20,2),
    'min_samples_leaf': np.arange(2,10,2),
    'max_features': ['None',2,4,6,8,10]
}

#Support Vector Machine
params_ran_sv = {
    'C': np.logspace(-2, 1, 10),
    'gamma':['scale',1,0.1,0.01,0.001,0.0001],
    'kernel':['rbf']}

#Logistic Regression
params_ran_log = {
    'penalty': ['l1','l2','none'],
    'C': np.logspace(-2, 10, 20),
    'class_weight': ['None','balanced']
}


#K neighbours
params_ran_knn = {
    'n_neighbors': np.arange(2,12,2),
    'weights': ['distance','uniform'],
    'p': [1,2]
}
```

```python
# Parameter Grid for Assemble model

# Stacking (Voting)
clf1 = LogisticRegression(random_state=42)
clf2 = RandomForestClassifier(random_state=42)
clf3 = MultinomialNB()
eclf = VotingClassifier(        
     estimators=[('lr', clf1), ('rf', clf2), ('mnb', clf3)],
     voting='soft'
)

params_ran_vote = {'lr__C': [2,3,4], 
          'rf__n_estimators': [100, 200],
          'rf__max_depth': [3, 4],
          'rf__max_features': [8,10,12],
          'mnb__alpha': [0.6,0.7,0.8 ]}

# Bagging
params_ran_bag = {'max_samples': [0.4,0.5,0.6], 
          'max_features': [0.4,0.5,0.6],
          'bootstrap': ['True','False'],
           }

# ADABoost
params_ran_ada= {'n_estimators': [50,100,150],  
             'base_estimator__max_depth': [2,3,4,5],
             'learning_rate': [.8, 1.,1.2],
           }

# RandomForest
params_ran_rf = {
    'n_estimators': np.arange(10,200,10),
    'max_depth' : np.arange(1,6,1),
    'min_samples_split': np.arange(2,20,2),
    'min_samples_leaf': np.arange(2,10,2),
    'max_features': ['None',2,4,6,8,10]
}

# GradientBoost
params_ran_gboost = {
    'max_depth': [2,3,4,5],
    'n_estimators': [50, 100, 150],
    'learning_rate': [.08, .1, .12]
}
```

```python
# Function defined = for RandomizedSearch - 1000 iteration and cross validation 3 fold
# Result storing dictionary
ran_result = {}

def random_gridsearch(model,parameter,name): 
  cv=3
  preprocess_step = None
    
  t0 = time.time()
  
  #Countvectorizer and Tfidf Vectorizer is processed with columntransformer - a list/array is passed instead of dataframe
  preprocessor_cvec = ColumnTransformer([("cvec", CountVectorizer(),'title_combined')])
  preprocessor_tfi = ColumnTransformer([("tfidf", TfidfVectorizer(),'title_combined')])

  #Pipe line for this section: preprocess -> classification model
  pipe_ran = Pipeline([('preprocess',preprocessor_cvec),("classifier", model)])
  
  #Parameter grid for each preprocess method + the model input
  #Countvectorizer
  param1 = {}
  param1['preprocess'] = [preprocessor_cvec]
  param1['preprocess__cvec__max_features'] = np.arange(1500,3000,500)
  param1['preprocess__cvec__min_df'] = [1,2,3]
  param1['preprocess__cvec__max_df'] = [0.85,0.95,1]
  param1['preprocess__cvec__ngram_range'] = [(1,1), (1,2), (1,3)]
  for key,value in parameter.items():
    param1['classifier'+"__"+key] = value

  #Tfidf vectorizer
  param2 = {}
  param2['preprocess'] = [preprocessor_tfi]
  param2['preprocess__tfidf__max_features'] = np.arange(1500,3000,500)
  param2['preprocess__tfidf__min_df'] = [1,2,3]
  param2['preprocess__tfidf__max_df'] = [0.85,0.95,1]
  param2['preprocess__tfidf__ngram_range'] = [(1,1), (1,2), (1,3)]
  for key,value in parameter.items():
    param2['classifier'+"__"+key] = value
  
  #Combine parameter grid
  params = []
  params = [param1, param2] 
    
  # run RandomsearchCV, scoring method :Accuracy
  gs_rand = RandomizedSearchCV(pipe_ran, params, cv=cv, scoring='accuracy',n_iter=1000, n_jobs=-1,verbose=1,random_state=42)
    
  # fit model
  gs_rand.fit(X_train, y_train)
  
  # Figure out which preprocess step is the best match with corresponding model
  if 'tfidf' in str(gs_rand.best_params_['preprocess']):
    preprocess_step = 'tfvec'
  elif 'cvec' in str(gs_rand.best_params_['preprocess']):
    preprocess_step = 'cvec'
  
  #Extracting all important information for the parameter with best score   
  model = gs_rand.best_estimator_
  pred = model.predict(X_test)
  probs = model.predict_proba(X_test)
  probs = probs[:, 1]
  auc = roc_auc_score(y_test, probs)
  fpr, tpr, _ = roc_curve(y_test, probs)
  tn, fp, fn, tp = confusion_matrix(y_test,pred).ravel()
  accuracy = (tp + tn) / (tp + tn + fp + fn)
  misclassification = 1 - accuracy
  sensitivity = tp / (tp + fn)
  specificity = tn / (tn + fp)
  precision = tp / (tp + fp)
  
  # Calculate the total original number of iteration if use gridsearchCV
  num_fit = cv*len(ParameterGrid(params))

  runtime = time.time() - t0
  
  #storing of result
  ran_result[name] = [preprocess_step, model.score(X_train, y_train), model.score(X_test, y_test),accuracy,auc,
                           misclassification, sensitivity, specificity, precision, runtime,fpr,tpr,num_fit]

  print(model) 
```

### Final Tuning of models: GridSearchCV

The parameter grid search range is already reduced by large amount by using the initial guess from randomized search result.

On the other hand, since all of them are using Count Vectorizer, the preprocessing steps is fix with minumum serach in this session

```python
# Parameter Grid for GridSearchCV 

#Random Forest
params_grid_rf = {
    'n_estimators': [180,190,200,210],
    'max_depth' : [4,5,6],
    'min_samples_split': [6,8,10],
    'min_samples_leaf': [2,4],
    'max_features': [8,10,12]
}

#Multinomial NB
params_grid_nb = {
    'alpha':[0.9,0.8,0.7,0.6,0.5,0.4]
    
}

#Gradient Boosting
params_grid_gboost = {
    'max_depth': [3,4,5],
    'n_estimators': [200,250,300],
    'learning_rate': [.1, .12],
    'min_samples_split': [6,8,10],
    'min_samples_leaf': [2,4],
    'max_depth' : [4,5,6],
    'max_features': [10,12]
}
```

```python
grid_result = {}

def full_gridsearch(model,parameter,name): 
  cv=3
  t0 = time.time()
  
  #Define preprocessing steps and pipeline
  preprocessor_cvec = ColumnTransformer([("cvec", CountVectorizer(),'title_combined')])
  pipe_grid = Pipeline([('preprocess',preprocessor_cvec),("classifier", model)])
  
  #Parameter grid setup
  param = {}
  param['preprocess__cvec__max_features'] = [1500,2000,2500]
  param['preprocess__cvec__max_df'] = [0.85,0.95]
  for key,value in parameter.items():
    param['classifier'+"__"+key] = value

  # run GridSearchCV and fitting
  gs = GridSearchCV(pipe_grid, param, cv=cv, scoring='accuracy', n_jobs=-1,verbose=1)
  gs.fit(X_train, y_train)
  
  #Models evaluation
  model = gs.best_estimator_
  pred = model.predict(X_test)
  probs = model.predict_proba(X_test)
  probs = probs[:, 1]
  auc = roc_auc_score(y_test, probs)
  fpr, tpr, _ = roc_curve(y_test, probs)
  tn, fp, fn, tp = confusion_matrix(y_test,pred).ravel() 
  accuracy = (tp + tn) / (tp + tn + fp + fn)
  misclassification = 1 - accuracy
  sensitivity = tp / (tp + fn)
  specificity = tn / (tn + fp)
  precision = tp / (tp + fp)

  runtime = time.time() - t0

  grid_result[name] = [model.score(X_train, y_train), model.score(X_test, y_test),accuracy,auc,
                           misclassification, sensitivity, specificity, precision, runtime,fpr,tpr]
  print(model)
```

```python
full_gridsearch(MultinomialNB(),params_grid_nb,'multinomial_Naive_Bayes')
```

```python
full_gridsearch(RandomForestClassifier(random_state=42),params_grid_rf,'random_forest')
```

```python
full_gridsearch(GradientBoostingClassifier(random_state=42),params_grid_gboost,'gradient_boosting')
```

##### Gradient boosting can be further tuned to close the generalisation by sacrificing small amount of accuracy

Method 1: Shrinkage/Weighted Updates
- A "learning rate" or 'number of features' is adjusted to prevent overfitting, depending on the structure of the decision tree (There's a trade-off between the learning rate and the number of trees needed). Since our learning rate already close to minimum(0.1), we will tuning the number of estimator here.

Method 2: Random Sampling/Stochastic Boosting
- In sklearn gradient boosting, we can subsampling columns before each split by using 'subsample' parameter. This technique essentially reduces the strength of the correlation between trees and prevent overfitting.


```python
#Gradient Boosting further tuning 
numner_estimator = np.arange(10,300,20)
```

```python
#Fix learning rate of 0.12
n_estimator_generalisation = []
n_estimator_score = []
n_estimator_AUC = []
for i in range (len(numner_estimator)):
    
  #Hyperparameter and pipeline setup
  preprocessor_cvec = ColumnTransformer([("cvec", CountVectorizer(max_df=0.85,max_features=1500),'title_combined')])
  pipe = Pipeline([('preprocess',preprocessor_cvec),
                   ("GBoost", GradientBoostingClassifier(learning_rate=0.12, max_depth=4,
                                            max_features=12, min_samples_leaf=2,
                                            min_samples_split=10,
                                            n_estimators=numner_estimator[i],
                                            random_state=42))])
  #fitting model
  pipe.fit(X_train, y_train)
  
  #Models evaluation
  train =  pipe.score(X_train, y_train)
  test = pipe.score(X_test, y_test)
  generalisation = abs((test-train)/test)*100
 
  n_estimator_score.append(train)
  n_estimator_generalisation.append(generalisation)
  
  probs = pipe.predict_proba(X_test)
  probs = probs[:, 1]
  auc = roc_auc_score(y_test, probs)
    
  n_estimator_AUC.append(auc) 
  
```

```python
num_df = pd.DataFrame(data={'num_estimator':numner_estimator , 'generalisation':n_estimator_generalisation,'score':n_estimator_score})
```

```python
fit, ax = plt.subplots(nrows = 2, ncols = 1, figsize=(12,10), sharex = False)
sns.lineplot(data=num_df, y='generalisation',  x ='num_estimator',ax=ax[0], markers=True, dashes=False,linewidth=3,color='indigo')
ax[0].set_title('Generalisation with number of features in Gboost',fontsize=20)
ax[0].set_xlabel('number of estimator', fontsize=15)
ax[0].set_ylabel("generalisation",fontsize=15)
ax[0].get_xaxis().set_visible(False)
ax[0].axhline(y=5, linestyle='--',color='indigo',label='generalisation target',linewidth=2)
ax[0].tick_params(axis='both', which='major', labelsize=12)
ax[0].grid
ax[0].legend(fontsize=15)

sns.lineplot(data=num_df, y='score', x ='num_estimator',ax=ax[1], markers=True, dashes=False,linewidth=3,color='skyblue')
ax[1].set_xlabel('number of estimator', fontsize=15)
ax[1].set_title('Score with number of features in Gboost',fontsize=18)
ax[1].set_ylabel("score",fontsize=15)
plt.subplots_adjust(hspace=0.09)

ax[1].tick_params(axis='both', which='major', labelsize=12)
```

By tuning the number estimator in gradient boost, we able to get a good generalisation score of below 5 % by maintaining the score of >90% which is robust enough for the prediction. (Number of Estimator between 120 to 150)

In conclusion, all the 3 models are having their own pros and cons:

1. Multinimial NB - Very fast computing speed with best train score and AUC, but less hyperparameter to tune and hard to close the generalisation gap

2. Random Forest - Has the lowest generalisation score from GridSearch, but having lower score and longest compiting time

3. Gradient Boosting - So far the best model, wide variaty of hyperparameter to tune, moderate computing time, able to achieve good score and generalisation gap
